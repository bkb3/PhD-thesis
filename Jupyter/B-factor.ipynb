```python
import numpy as np
import seaborn as sns
from matplotlib import cm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.4)
sns.set_palette("Set1")
```

```python
def asf(B, y):
    return np.exp(-B*y**2)
```

```python
b_10 = [asf(10, i) for i in np.linspace(0, 0.5)]
b_20 = [asf(20, i) for i in np.linspace(0, 0.5)]
b_30 = [asf(30, i) for i in np.linspace(0, 0.5)]
```

```python
plt.figure(figsize=(4,4))
sns.lineplot(np.linspace(0, 0.5), b_10, label='$B = 10 \AA^2$')
sns.lineplot(np.linspace(0, 0.5), b_20, label='$B = 20 \AA^2$')
ax = sns.lineplot(np.linspace(0, 0.5), b_30, label='$B = 30 \AA^2$')
ax.set_xlabel('$\\frac {sin \\enspace \\theta}{\lambda}$')
ax.set_ylabel('$f_B$')
plt.savefig('figs/bfactors.pdf', bbox_inches = 'tight', pad_inches = 0)
```
